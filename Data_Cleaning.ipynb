```python
import pandas as pd
import numpy as npy
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Reading the csv file
df = pd.read_csv('movies.csv')
df
```

```python
#Checking the dataset for missing values
df.isnull().sum()
```

```python
#Replacing all null values with 0
df.fillna(0, inplace=True)
df
```

```python
#Checking to see that there are no more missing values
df.isnull().sum()
```

```python
#Checking for duplicates
df.duplicated()
```

```python
#Ensuring correct data types of columns
df.dtypes
```

```python
#Ensuring our dataset is fit for analysis
#dropping all duplicate values
df.drop_duplicates(inplace= True)
df
```

```python
#Checking to see that there are no duplicate values
df.duplicated().sum()
```

```python
#Renaming a column
df.rename(columns = {"STARS" : "STARS AND DIRECTORS"}) 
```

```python
#since there are no more missing values and duplicates
print('My data set fit for analysis')
```

###I read the movies.csv dataset into my dataframe and checked for missing values which were replaced with '0'. I also checked for duplicate values which were all dropped after they were discovered. I also renamed the 'STARS' column name since I noticed that it did not only carry the name of the movie stars but also the movie directors.
