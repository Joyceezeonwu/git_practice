```python
#Importing Libraries
import pandas as pd
import numpy as npy
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the tips dataset
df= sns.load_dataset('tips')

df
```

```python
#Generating a KDE plot for the size feature of the tips dataset
plt.figure(figsize=(8,6))
sns.kdeplot(data=df,shade=True)
plt.title('Size Feature of the Tips Dataset')
plt.legend(['total_bill', 'tip', 'size'])
plt.savefig('Size Feature of the Tips Dataset.png')
plt.show()
```

####The size and amount of tips are collected from the dataset and the total bills also generated.

```python
#Generating the distribution of total bill by gender for the tips dataset using the boxplots. 
plt.figure(figsize=(8,6))
sns.set(style="darkgrid")
my_pal = {"Male" : "b", "Female" : "m"}
sns.boxplot(data=df, x=df["sex"], y=df["total_bill"], palette=my_pal)
plt.title('Total Bill Distribution by Gender')
plt.savefig('Total Bill Distribution by Gender.png')
plt.show()
```

####It is evident here that the there were more bills generated by male, with a max. value of approximately 40 which is higher than the female folk of 30


```python
#Generating the distribution of tips by days using boxplots
plt.figure(figsize=(8,6))
sns.set(style="darkgrid")
my_pal = {"Thur" : "b", "Fri" : "m", "Sat" : "g", "Sun" : "r"}
sns.boxplot(data=df, x=df["day"], y=df["tip"], palette= my_pal)
plt.title('Distribution of Tips by Days')
plt.savefig('Distribution of Tips by Days.png')
plt.show()
```

####From the plot above, Sunday had a higher generation of tips with Thursday and Saturday having same value and Friday generating the least tips.

```python
#Generating the distribution of total bill by days considering smoker status of the customer using box plots
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='day', y='total_bill', hue='smoker')
plt.title('Distribution of Total Bill by Day based on Smokers Status')
plt.savefig('Distribution of Total Bill by Day based on Smokers Status.png')
plt.show()
```

####From the plot above, Sunday had a higher number of smokers, followed by Saturday and then Thursday while Friday had the least number of smokers with more people who smoked. 

```python
#Generating the distribution of total bill by time given the genders using violin plots. 
plt.figure(figsize=(8,6))
sns.set(style="darkgrid")
my_pal = {"Male" : "b", "Female" : "m"}
sns.violinplot(data=df, y='time', x='total_bill', hue='sex', orient='horizontal', width=0.5, palette=my_pal)
plt.title('Distribution of Total Bill by Time depending Gender')
plt.savefig('Distribution of Total Bill by Time depending Gender.png')
plt.show()
```

####During lunch more females had bills to pay, however the difference is not much compared to dinner where total bills generated by the males showed a larger difference.

```python
#Generating the distribution of tips by days also considering time using violin plots 
plt.figure(figsize=(8,6))
sns.set(style="darkgrid")
my_pal = {"Lunch" : "b", "Dinner" : "y"}
sns.violinplot(data=df, y='day', x='tip', hue='time', orient='horizontal', width=0.5, palette=my_pal)
plt.title('Distribution of Tips by Days depending on Time')
plt.savefig('Distribution of Tips by Days depending on Time.png')
plt.show()
```

####On Thursday, there were no tips generated at Dinner while Saturday and Sunday saw no tips generated at Lunch time. Friday had tips generated at Lunch and Dinner with more tips collected during dinner.

```python
#Plotting and customizing a correlations heatmap for the tips dataset
plt.figure(figsize=(8,6))
ax = sns.heatmap(df.corr(), annot=True, fmt='.1f', cmap='Oranges', linewidths=9, cbar_kws={'shrink':.99})
plt.title('Correlations Heatmap of Tips Dataset')
plt.savefig('Correlations Heatmap of Tips Dataset.png')
plt.show()
```
